```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("thedevastator/books-sales-and-ratings")

print("Path to dataset files:", path)
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title ("plt DISTRIBUITION")
plt.show()
```

```python
df["genre"].value_counts().plot(kind = "bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = True).head()
```

```python
sns.boxplot(x="genre",y="Book_average_rating" , data =df)
plt.xlabel("Genre")
plt.ylabel("Books rating type shit")
plt.title("this is a box plot -_-")
plt.show()
```

SCATTER PLOT FOR FINDING THE PRICE TO UNITS SOLD


```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Books sales type shit")
plt.title("this is a SCATTER plot T-T")
plt.show()
```
